## Section 3.2: Continunity and Differentiation

#### Section 3.2.1 Limits and Continuity

A function is defined as **continuous** if it does not have any discontinuities.

Continuous function definition: Let $ f:B⟶ ℝ $ be a real-valued function on $ B ⊆ ℝ^d $. Then $f$ is said to be continuous at $a\in B$ if $ lim_{x→a} f(x) = f(a)$.

An example of a continuous function is dhown below:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def polynomial(x):
    return x**2 + 5*x + 3

# Values for x
x = np.linspace(-10, 10, 400)

# Calculate corresponding y values
y = polynomial(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y,)

plt.title('Plot of Polynomial Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Continuity is maintained under composition. Meaning, if $f$ is continuous at $x_0$ and $g$ is continuous $f(x_0)$, then $g ∘f$ is continuous at $x_0$.

The **extremum lemma** states that if a function is continuous on a closed interval, then somewhere in the interval, its maximum and minimum point is located!



#### Section 3.2.2 Derivatives

Definition of **Derivative**: Let $f:D⟶ℝ $, where $D⊆ℝ$ and let $x_0 \in D$ be a point inside D. Then, the derivative of $f$ at $x_0$ is

$f'(x_0) = \frac{df(x_0)}{dx} = lim_{h→0} \frac{f(x_0 +h) -f(x_0) }{h}$,

given that the limit exists.

\
**Lemma**: Let $f$ and $g$ have derivatives at $x$ and let $α$ and $β$ be constants. Then, the following is true

$ [αf(x)+βg(x)]' = αf'(x) + βg'(x)$ .

This property allows us to calculate the extreme values of the functions!

\
Below is an example of how to calculate for derivatives in Python:

```python
import sympy as sp

# Define the variable
x = sp.symbols('x')

# Define the  function
f = x**2 + 5*x + 3

# Calculates the derivative
derivative = sp.diff(polynomial, x)

# Display the derivative
print(f"The derivative of the function {f} is: {derivative}")
```

**Rolle Theorem** states that if $f : [a,b] ⟶ ℝ$ is continuous, its derivative exists on $(a,b)$, and $f(a)=f(b)$, then there exists a point $c$ such that $a<c<b$ and $f'(c)=0$.

\
The **Mean Value Theorem** states that if $f: [a,b] ⟶ ℝ$ is continuous and its derivative exists on $(a,b)$, then there exists point $c$ such that $a<c<b$ and $\frac{f(b)-f(a)}{b-a} =f'(c)$.

\
The **partial derivative** of $f:D⟶ℝ$ with respect to $x_i$, where $x⃗ _0 \in D$ is

$\frac{∂f(x⃗ _0)}{∂x_i} = lim_{h→0} \frac{f(x⃗_0+he⃗ _i)-f(x⃗_0)}{h}$,

where $e⃗ _i \in ℝ^d$ is the i-th standard basis vector.

\
Definition of **Jacobian**: Let $f⃗=(f_1, ... , f_m) : D⟶ ℝ^m$ where $D ⊆ℝ^d$ and $x⃗ _0$ is an interior point of $D$ where $\frac{∂f_j(x⃗ _0)}{∂x_i}$ exists for all $i,j$. Then, the Jacobian of $f⃗$ at $x⃗ _0$ is the $d×m$ matrix

$J_f(x⃗_0) =$ \begin{pmatrix}
            \frac{∂f_1(x⃗ _0)}{∂x_1} & ⋯ & \frac{∂f_1(x⃗ _0)}{∂x_d} \\
            \vdots & \ddots & \vdots \\
            \frac{∂f_m(x⃗ _0)}{∂x_1} & ⋯ & \frac{∂f_m(x⃗ _0)}{∂x_d}
            \end{pmatrix}

\
or reduced in vector form:

$Jf(x⃗_0) = ∇f(x⃗_0)^T$,

\
where

$∇ f(x⃗_0) = (\frac{∂f(x⃗ _0)}{∂x_1}, ... , \frac{∂f(x⃗ _0)}{∂x_d})^T$.

\
Assuming that $f$ is continuously differentiable at $x_0$ and g is continuously differentiable at $f(x_0)$, the **chain rule** states

$J_{g∘f}(x_0) = J_g(f(x_0))J_f(x_0)$

as a product of matrices!

\
The **directional derivative** of $f$ at $x_0$ in the direction of the unit vector $v \in ℝ^d$ is

$\frac{∂f(x⃗ _0)}{∂v⃗} = lim_{h→0}\frac{f(x⃗_0+hv⃗)-f(x⃗_0)}{h}$,

given that the limit exists.

\
Assuming that $f$ is continuously differentiable in an open ball around $x⃗_0$, then the **second partial derivative** in respect to $x_j$ (given that it exists) is denoted by

$\frac{∂^2f(x⃗_0)}{∂x_j∂x_i} = lim_{h→0} \frac{∂f(x⃗_0 + he⃗_j)/∂x_i-∂f(x⃗_0)/ ∂x_i}{h}$.

\
The Jacobian of the gradient $∇f$ is referred to as the **Hessian** and is denoted by

$H_f(x⃗_0) =$ \begin{pmatrix}
            \frac{∂^2f(x⃗_0)}{∂x_1 ^2} & ⋯ & \frac{∂^2f(x⃗ _0)}{∂x_d∂x_1} \\
            \vdots & \ddots & \vdots \\
            \frac{∂^2f(x⃗ _0)}{∂x_1∂x_d} & ⋯ & \frac{∂^2f(x⃗ _0)}{∂x^2_d}
            \end{pmatrix}


#### Section 3.2.3 Taylor's Theorem

**Taylor's theorem** allows us to approximate a differentiable function with a polynomial. Note" The greater the polynomial degree, the more accurate our approximation is.

\
Taylor's theorem states that for some differentiable function $f$ over $[a,b]$,

$f(b) =f(a) +(b-a)f'(a) +\frac{1}{2} (b-a)^2f''(a) + ... + \frac{(b-a)^{m-1}}{(m-1)!}f^{m-1}(a)+R_m$,

where $R_m= \frac{(b-a)^m}{m!}f^m(a+θ(b-a))$ for some $0<θ<1$.

\
Below is an example on using taylor series to approximate a function in Python (code sampled from Copilot):

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the variable and the function
x = sp.symbols('x')
y = sp.exp(x)

# Define the Taylor series expansion
taylor_series = sp.series(y, x, 0, 5).removeO()

# Convert the Taylor series to a lambda function, which will be used for plotting
taylor_series_func = sp.lambdify(x, taylor_series, 'numpy')

# Provide x values
x_vals = np.linspace(-2, 2, 400)

# Calculate y values for the original function and the Taylor series approximation
y_vals_original = np.exp(x_vals)
y_vals_taylor = taylor_series_func(x_vals)

# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals_original, label=r'$e^x$', color='blue')
plt.plot(x_vals, y_vals_taylor, label=r'Approximation', color='red', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Taylor Series Approximation of $e^x$')
plt.legend()
plt.show()
```

#### Section 3.3.1 Necessary and Sufficient Conditions of Local Minimizers

Let $f:ℝ^d ⟶ℝ$. Then, the point $x^* \in ℝ^d$ is a **global minimizer** of $f$ if

$f(x)≥f(x^*), ∀x \in ℝ^d$.

\
With the same conditions as before, if there is $δ>0$ such that

$f(x)≥ f(x^*), ∀x ∈ B_δ(x^*)$ \ {$x^*$},

then the point $x^* \in ℝ^d$ is a **local minimizer** of $f$.

Below is an example of how to find global and local minimizers in Python:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Defines the function
def f(x):
    return 3*x**4 - 10*x**2 + x +1

# Creates x values
x_values = np.linspace(-2, 2, 100)
y_values = f(x_values)

# Plot the function
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Global and Local Minimizers of f(x) = 3x^4 - 10x^2 + x + 1')

# Finds the global minimum (starts at 0)
result_global = minimize(f, x0=0)
x_global_min = result_global.x
y_global_min = result_global.fun

# Plots the global minimum
plt.plot(x_global_min, y_global_min, 'mo', label=f'Global Minimizer: ({x_global_min[0]:.2f}, {y_global_min:.2f})')

# Finds local minimum (starts at 1)
result_local = minimize(f, x0=1)
x_local_min = result_local.x
y_local_min = result_local.fun

# Plots the local minimum
plt.plot(x_local_min, y_local_min, 'bo', label=f'Local Minimizer: ({x_local_min[0]:.2f}, {y_local_min:.2f})')
plt.legend()
plt.show()
```

Assuming $f: ℝ^d ⟶ ℝ$ and that $α^*>0$ exists such that

$f(x⃗_0 +αv⃗)<f(x⃗_0), ∀α \in (0,α^*)$,

then vector $v⃗$ is a **descent direction** for f at $x⃗_0$.

\
The **existence of a descent direction** theorem states that a continuously differentiable function $f$ at $x⃗_0$ has a descent direction at $x⃗_0$ if $∇f(x⃗_0) \neq 0$.

\
The **first-order necessary condition** states that, assuming $f$ is continuously differentiable on $ℝ^d$, if $x⃗_0$ is a local minimizer, then $∇f(x⃗_0)=0$.

\
The **second-order necessary condition** states that if a twice continuously differentiable function $f$ on $ℝ$ has a local minimizer $x⃗_0$, then the Hessian of the function at $x⃗_0$ is positive semi-definite (PSD).

\
The **second-order sufficient condition** states that if $f$ is twice continuously differentiable on $ℝ^d$, $∇f(x⃗_0)=0$, and the Hessian of the function at $x⃗_0$ is positive definite, then $x⃗_0$ is a strict local minimizer.

#### Section 3.3.2 Convexity and Global Minimizers

A function is referred to as **convex** if its shape curves upwards (like that of a parabola, $x^2$) or, in other words, if the line segment between any two points in its domain lies above the graph between the two points.

\
**Convex Set** Definition: A set $A ⊆ ℝ^d$ is convex if for all $x⃗,y⃗ \in A$ and all $α \in [0,1]$

$(1-α)x⃗ + αy⃗ \in A$.

**Convex Function** Definition: A function $f: ℝ^d ⟶ ℝ$ is convex if for al $x⃗,y⃗ \in ℝ^d$ and all $α \in [0,1]$

$f((1-α)x⃗+αy⃗) ≤ (1-α)f(x⃗)+ αf(y⃗)$.

\
**First-Order Convexity Condition** Theorem: Let $f: ℝ^d ⟶ ℝ$ be continuously differentiable. Then $f$ is convex if and only if

$f(y⃗) ≥ f(x⃗) + ∇f(x⃗)^T(y⃗-x⃗) $

for all $x⃗,y⃗ \in ℝ^d$.

\
**Second-Order Convexity Condition** Theorem: Let $f: ℝ^d ⟶ ℝ$ be twice continuously differentiable. Then $F$ is convex if and only if $H_f(x⃗)$ is PSD for all $x⃗ \in ℝ^d$.

\
**Global Minimizers of Convex Functions** Theorem: Let $f: ℝ^d ⟶ ℝ$ be a convex function. Then any local minimizer of $f$ is also a global minimizer.

#### Section 3.3.3 Gradient Descent

To identify a global minimizer of $f$, it is best to find all $x⃗$'s such that $∇f(x⃗)=0$ and then choose the $x⃗$ that derives the smallest value of $f(x⃗)$.

\
**Steepest Descent** Theorem: Let $f: ℝ^d ⟶ ℝ$ be continuously differentiable at $x⃗
_0$. Then for any unit vector $v⃗ \in ℝ^d$,

$\frac{∂f(x⃗ _0)}{∂v⃗} ≥ \frac{∂f(x⃗ _0)}{∂v⃗^*}$,

where $v⃗^* = - \frac{∇f(x⃗_0)}{||∇f(x⃗_0)||}$.

\
The stepsize, $α_k>0$, plays a crucial role in reaching the global minimizer. If it is too big, then the global minimum will likely not be found. Although a smaller stepsize gets you closer to the actual global minimizer, it is also still not a guarantee and it may never be reached. In sum, the final iteration of the steepest descent is not guaranteed to be the global minimizer.
